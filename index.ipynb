```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline

iris = load_iris()
target_names = np.take(iris.target_names, iris.target)
data = np.concatenate(
    (iris.data, np.array([iris.target]).T, np.array([target_names]).T), 
    axis=1
)

df = pd.DataFrame(
    data=data, 
    columns=iris.feature_names + ['target', 'target_name']
)
df = df.apply(pd.to_numeric, errors='ignore')


formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])

x_col_label = 'petal length (cm)'
y_col_label = 'petal width (cm)'

plt.scatter(df[x_col_label], df[y_col_label], c=df['target'])
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(x_col_label)
plt.ylabel(y_col_label)

plt.tight_layout()
plt.savefig('slide_5_graph.png', dpi=500)
plt.show()
```

```python
from pandas.plotting import table

# df.head(10)

ax = plt.subplot(111, frame_on=False)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)

table(ax, df[:10], loc='center')

plt.savefig('slide_4_table.png', dpi=500)
```

# Scratches

```python
x1, y1 = [1, 2, 3, 4], [10, 20, 30, 40]
x2, y2 = [2, 3, 4, 5], [10, 20, 30, 40]

# Using matplotlib instances
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x1, y1)
ax.scatter(x2, y2, color='red')
ax.set_xlim(0, 5.5)
plt.show()

# Using matplotlib.pyplot functions
plt.plot(x1, y1)
plt.scatter(x2, y2, color='red')
plt.xlim(0, 5.5)
plt.savefig('1_graph.png', dpi=500)
plt.show()
```

```python
INITIAL_X, INITIAL_Y = 0, 0
WIDTH, HEIGHT = 1, 1
fig = plt.figure()
ax = fig.add_axes((INITIAL_X, INITIAL_Y, WIDTH, HEIGHT))
ax.plot([1, 2], [3, 4])
plt.savefig('2_graph.png', dpi=500)
```

```python
ROWS_NUMBER = 1
COLS_NUMBER = 1

fig = plt.figure()
ax = fig.add_subplot(ROWS_NUMBER, COLS_NUMBER, 1)
ax.plot([1, 2], [3, 4])
plt.savefig('3_graph.png', dpi=500)
```
