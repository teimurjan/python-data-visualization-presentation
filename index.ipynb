```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline

iris = load_iris()
target_names = np.take(iris.target_names, iris.target)
data = np.concatenate(
    (iris.data, np.array([iris.target]).T, np.array([target_names]).T), 
    axis=1
)

df = pd.DataFrame(
    data=data, 
    columns=iris.feature_names + ['target', 'target_name']
)
df = df.apply(pd.to_numeric, errors='ignore')


formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])

x_col_label = 'petal length (cm)'
y_col_label = 'petal width (cm)'

plt.scatter(df[x_col_label], df[y_col_label], c=df['target'])
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(x_col_label)
plt.ylabel(y_col_label)

plt.tight_layout()
plt.savefig('1.png')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x1, y1 = [1, 2, 3, 4], [10, 20, 30, 40]
ax.plot(x1, y1, c="red")
fig.savefig('2.png')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

x1, y1 = [1, 2, 3, 4], [5, 7, 9, 11]
ax.plot(x1, y1, c="darkblue", linestyle="--")

ax.set_xlabel('I am xlabel')
ax.set_ylabel('I am ylabel')
ax.set_title('I am title')
ax.legend(
    ['Straight line with some meaning'], 
    loc="upper center"
)
ax.set_xticks(x1)
ax.set_yticks(y1)

fig.tight_layout()
fig.savefig('3.png', transparent=True)
```
