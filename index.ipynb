```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def get_df_from_iris(iris):
    target_names = np.take(iris.target_names, iris.target)
    data = np.concatenate(
        (iris.data, np.array([target_names]).T), 
        axis=1
    )

    df = pd.DataFrame(
        data=data, 
        columns=iris.feature_names + [ 'target_name']
    )
    return df.apply(pd.to_numeric, errors='ignore')

iris = load_iris()
df = get_df_from_iris(iris)
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])

x_col_label = 'petal length (cm)'
y_col_label = 'petal width (cm)'

plt.scatter(df[x_col_label], df[y_col_label], c=iris.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(x_col_label)
plt.ylabel(y_col_label)

plt.tight_layout()
plt.savefig('slide_5_graph.png', dpi=500)
plt.show()
```

```python
from pandas.plotting import table

ax = plt.subplot(111, frame_on=False)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)

table(ax, df.sample(10, random_state=1), loc='center')

plt.savefig('slide_4_table.png', dpi=500, bbox_inches='tight')
```

```python
iris = load_iris()
iris_df = get_df_from_iris(iris)

fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
col_label = 'petal length (cm)'
is_setosa = iris_df['target_name'] == 'setosa'
col = iris_df[is_setosa][[col_label]]
data = np.hstack(col.values)
ax.set_xlabel(col_label)
ax.set_ylabel('quantity')
ax.hist(data)


plt.savefig('scratches/10_graph.png', dpi=500)
```
